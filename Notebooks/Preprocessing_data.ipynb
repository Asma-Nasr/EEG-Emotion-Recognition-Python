## Necessary libraries

```python
import pandas as pd
import numpy as np
from src.filter import SignalFilter
import matplotlib.pyplot as plt
```

## Loading data

```python
calm = pd.read_excel('Data/CALM.xlsx')
fear = pd.read_excel('Data/FEAR.xlsx')
happy = pd.read_excel('Data/Happy.xlsx')
sad = pd.read_excel('Data/SAD.xlsx')
```

```python
print("Calm shape",calm.shape)
print("fear shape",fear.shape)
print("happy shape",happy.shape)
print("sad shape",sad.shape)
```

## Filtering

```python
fs = 256
num_channels = 19  # Number of channels/features
filter_type = 'low'  # Choose 'low', 'high', or 'band'
signal_filter = SignalFilter(sample_frequency=fs, filter_type=filter_type)
```

## Calm

```python
filtered_calm = signal_filter.apply_filter(calm) 
filtered_calm = np.array(filtered_calm)[0]
```

```python
# Plot the results
num_samples = calm.shape[0]  # Number of samples/records
t = np.linspace(0, num_samples / fs, num_samples, endpoint=False)
plt.figure(figsize=(12, 6))

# Original Signal
plt.subplot(2, 1, 1)
plt.title('Original calm Signal')
plt.plot(t, calm)  

# Filtered Signal
plt.subplot(2, 1, 2)
plt.title('Filtered calm Signal')
plt.plot(t, filtered_calm)
plt.tight_layout()
plt.show()
```
