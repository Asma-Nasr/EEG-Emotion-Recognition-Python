## Necessary libraries

```python
import pandas as pd
import numpy as np
from src.mfcc import extract_and_save_mfcc_from_folder
import matplotlib.pyplot as plt
```

```python
extract_and_save_mfcc_from_folder('Data/Filtered Data', 'Data/Feature Extracted')
```

## Loading data

```python
calm = np.load('Data\Feature Extracted\calm_mfcc.npy')
fear = np.load('Data\Feature Extracted/fear_mfcc.npy')
happy = np.load('Data\Feature Extracted/happy_mfcc.npy')
sad = np.load('Data/Feature Extracted/sad_mfcc.npy')
```

```python
print("Calm shape",calm.shape) #236 records 228 features
print("fear shape",fear.shape)
print("happy shape",happy.shape)
print("sad shape",sad.shape)
```

```python
num_samples = calm.shape[0]  # Number of samples/records
fs = 256
t = np.linspace(0, num_samples / fs, num_samples, endpoint=False)
```

## Visualization of the MFCC features across the 4-Classes

```python
plt.figure(figsize=(14,8))
plt.subplot(2, 2, 1)

# Feature Extracted Signal
plt.title('MFCCed calm Signal')
plt.plot(t, calm)  
plt.subplot(2, 2, 2)

plt.title('MFCCed fear Signal')
plt.plot(t, fear)

plt.subplot(2, 2, 3)
plt.title('MFCCed happy Signal')
plt.plot(t, happy)  

plt.subplot(2, 2, 4)
plt.title('MFCCed sad Signal')
plt.plot(t, sad)  
plt.tight_layout()
plt.show()
```
